```python
driver = webdriver.Chrome(r'C:\Users\hp\Downloads\chromedriver_win32 (1)\chromedriver.exe')
driver.get("https://www.amazon.in/")
```

```python
search = driver.find_element_by_xpath('//*[@id="twotabsearchtextbox"]')
a=input("Enter user's input: ")   #User's input
search.send_keys(a)
```

```python
search1 =driver.find_element_by_xpath('//input[@id="nav-search-submit-button"]').click()        #search for the input value
time.sleep(2)
```

```python
url =[]

guitar_url = driver.find_elements_by_xpath('//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
for i in guitar_url:
    url.append(i.get_attribute('href'))

url
```

```python
a = len(url)
a
```

```python
brand =[]
product=[]
prices=[]
exchange=[]
deliveries=[]
availabilities=[]
pro_urls=[]

for i in url[0:5]:
    driver.get(i)
    time.sleep(1)
    try:
        brand_name = driver.find_element_by_xpath('//*[@id="bylineInfo_feature_div"]/div')
        brand.append(brand_name.text)
    except NoSuchElementException:
        brand.append("NA")
    try:
        Name_of_the_Product  = driver.find_element_by_xpath('//*[@id="productTitle"]')
        product.append(Name_of_the_Product.text)
    except NoSuchElementException:
        product.append("NA")
    try:
        price = driver.find_element_by_xpath('//*[@id="corePriceDisplay_desktop_feature_div"]/div[1]/span[2]')
        prices.append(price.text)
    except NoSuchElementException:
        prices.append("NA")
    try:
        return_exchange = driver.find_element_by_xpath(' //*[@id="RETURNS_POLICY"]/span/div[2]/a')
        exchange.append(return_exchange.text)
    except NoSuchElementException:
        exchange.append("NA")
    try:
        delivery = driver.find_element_by_xpath('//*[@id="mir-layout-DELIVERY_BLOCK-slot-DELIVERY_MESSAGE"]/b')
        deliveries.append(delivery.text)
    except NoSuchElementException:
        deliveries.append("NA")
    try:
        availability = driver.find_element_by_xpath('//*[@id="availability"]')
        availabilities.append(availability.text)
    except NoSuchElementException:
        availabilities.append("NA")
    try:
        pro_urls.append(url)
    except NoSuchElementException:
        pro_urls.append("NA")
   
```

```python
len(pro_urls)
```

```python
data = pd.DataFrame()
data['Brand-Name']=brand
data['Product Name']=product
data['Price']=prices
data['Return and Exchange']=exchange
data['Expected Delivery']=deliveries
data['Availability']=availabilities
data['Product Url']=pro_urls

data.head(20)
```
