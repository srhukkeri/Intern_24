```python
laptop_url= []

laptops = driver.find_elements_by_xpath('//div[@class="product-ickon"]')
for i in laptops:
    laptop_url.append(i.get_attribute('data-href'))
a=len(laptop_url)

print(laptop_url)
print(a)
```

```python
product_des =[]


for i in laptop_url[0:a]:
    driver.get(i)
    time.sleep(1)
    try:
        details=driver.find_element_by_xpath('/html/body/div[3]/div[2]')
        product_des.append(details.text)
    except NoSuchElementException:
        product_des("NA")
```
