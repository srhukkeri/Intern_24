#### intel core i9

```python
#importing all the necessary libraries  
import selenium 
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import time
import requests

driver = webdriver.Chrome(r'C:\Users\s.rajendra.hukkeri\Downloads\chromedriver_win32\chromedriver.exe')   # downladed chrome extension 
url= "https://www.amazon.in"                                                                 #url of the website
driver.get(url)
search = driver.find_element_by_xpath('//*[@id="twotabsearchtextbox"]')
search.send_keys("Laptop")                                                             #send the value
search1 =driver.find_element_by_xpath('//input[@id="nav-search-submit-button"]').click()        #search for the input value
time.sleep(2)

clk = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[1]/div[2]/div/div[3]/span/div[1]/div/div/div[6]/ul[3]/li[13]/span/a/span')
clk.click()

#creating a list and applying "for" loop
titles=[]
price=[]

title=driver.find_elements_by_xpath('//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]')
for i in title:
    titles.append(i.text)

prices=driver.find_elements_by_xpath('//span[@class="a-price"]')
for j in prices:
    price.append(j.text)

time.sleep(2)

#creating the dataframe 
data = pd.DataFrame()
data['Title']=titles
data['Price']=price

data
```
