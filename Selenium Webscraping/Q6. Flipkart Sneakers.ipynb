```python
import selenium 
from selenium import webdriver
import pandas as pd
import warnings
import requests
warnings.filterwarnings("ignore")
import time

driver = webdriver.Chrome(r'C:\Users\s.rajendra.hukkeri\Downloads\chromedriver_win32\chromedriver.exe')
driver = webdriver.Chrome('chromedriver.exe')
url= "https://www.flipkart.com/"
driver.get(url)
button = driver.find_element_by_xpath('/html/body/div[2]/div/div/button').click()
search=driver.find_element_by_class_name("_3704LK")
search.send_keys("sneakers")
search1=driver.find_element_by_class_name("L0Z3Pu").click()      # search
time.sleep(5)
```

```python
href_urls = [my_elem.get_attribute("href") for my_elem in driver.find_elements_by_css_selector("nav.yFHi8N > a")] 
brand=[]
des=[]
prices=[]
offers=[]

for i in range(0,3):
    page = requests.get(url)   #url page
    driver = webdriver.Chrome(r'C:\Users\s.rajendra.hukkeri\Downloads\chromedriver_win32\chromedriver.exe')
    url= href_urls[i]
    driver.get(url)
    time.sleep(2) 
    brand_name = driver.find_elements_by_xpath('//div[@class="_2WkVRV"]')
    for l in brand_name:
        brand.append(l.text)
        describe = driver.find_elements_by_xpath('//a[contains(@class, "IRpwTa") or contains(@class,"IRpwTa _2-ICcC")]')
    for j in describe:
        des.append(j.text)
        price = driver.find_elements_by_xpath('//div[@class="_30jeq3"]')
    for k in price:
        prices.append(k.text)
        offer = driver.find_elements_by_xpath('//div[@class="_3Ay6Sb"]')
    for m in offer:
        if m.text == ' ':
            offers.append("No offer found")
        else:
                offers.append(m.text)


df = pd.DataFrame()
df['Brand']= brand
df['Price']= prices
df['Offer']= offers
df['Describe']=des
df.head(100)
```
