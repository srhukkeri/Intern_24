```python
import selenium 
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
driver = webdriver.Chrome(r'C:\Users\s.rajendra.hukkeri\Downloads\chromedriver_win32\chromedriver.exe')
url= 'https://www.ambitionbox.com/'
driver.get(url)
salary = driver.find_element_by_xpath('//a[@title="Company Salaries"]').click()
```

```python
searchbox = driver.find_element_by_xpath('//input[@type="searchbox"]')
searchbox.send_keys("Data Scientist")
time.sleep(2)
select1 = driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div/p')
select1.click()
```

```python
companyname=[]
company = driver.find_elements_by_xpath('//div[@class="name"]')
for i in company:
    companyname.append(i.text)
companyname
```

```python
average_salary=[]
average = driver.find_elements_by_xpath('//p[@class="averageCtc"]')
for j in average:
    average_salary.append(j.text)
average_salary
```

```python
min_salary=[]
minimum = driver.find_elements_by_xpath('//div[@class="salary-values"]')
for k in minimum:
    min_salary.append(k.text)
min_salary
```

```python
experience=[]
exps = driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]')
for m in exps:
    experience.append(m.text)
experience
```

```python
data=pd.DataFrame()
data['Company Name']=companyname
data['Average Salary']=average_salary
data['Minimum and Maximum Salary']=min_salary
data['Experience']=experience
data
```

```python
data[['Minimum', 'Maximum']] = data['Minimum and Maximum Salary'].str.split('\n', 1, expand=True)
data
```

```python
data.drop(columns=['Minimum and Maximum Salary'], axis=1, inplace=True)
data
```

```python
data[['Exps','Experiences']] = data['Experience'].str.split('Data Scientist\n . \n', 1, expand=True)
data
```

```python
data.drop(columns=['Experience'], axis=1, inplace=True)
data
```

```python
data.drop(columns=['Exps'], axis=1, inplace=True)
data
```
